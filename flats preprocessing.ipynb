```python
def treat_price(x):
    if isinstance(x, list) and len(x) >= 2:  # Check if x is a list and has at least two elements
        if x[1] == 'lac' or x[1]=='Lac':
            return round(float(x[0])/100, 2)
        else:
            return round(float(x[0]), 2)
    else:
        return x
```

```python
df['price']=df['price'].str.split(' ').apply(treat_price)
```

```python
df
```

```python
df['price_per_sqft']=df['price_per_sqft'].apply(lambda name:re.sub('/sq\.ft\.','',str(name))).str.strip()
```

```python
df
```

```python
df = df[df['price_per_sqft'].str.replace('₹', '').str.replace(',', '').str.strip().str.isnumeric()]

df['price_per_sqft'] = df['price_per_sqft'].str.replace('₹', '').str.replace(',', '').str.strip().astype(float)
```

```python
df['bedRoom'].isnull().sum()
```

```python
df['bedRoom'].value_counts()
```

```python
df['bedRoom']=df['bedRoom'].str.replace('Bedrooms','')
df['bedRoom']=df['bedRoom'].str.replace('Bedroom','').astype('int')
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'] = df['bathroom'].str.replace('Bathrooms', '')
df['bathroom'] = df['bathroom'].str.replace('Bathroom', '').astype('int')
```

```python
df['balcony'].value_counts()
```

```python
df['balcony']=df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
df['floorNum'].value_counts()
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0)  # Extracting the first part of the split string
df['floorNum'] = df['floorNum'].replace('Ground', '0')  # Replacing 'Ground' with '0'
df['floorNum'] = df['floorNum'].str.replace('Basement', '-1')  # Replacing 'Basement' with '-1'
df['floorNum'] = df['floorNum'].str.lower()  # Converting to lowercase
df['floorNum'] = df['floorNum'].str.extract(r'(\d+)')  # Extracting digits
```
